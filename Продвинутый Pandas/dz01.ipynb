# Задание 1

```python
import pandas as pd
```

```python
visits = pd.read_csv('visit_log.csv', sep = ';')
visits.head()
```

```python
def source_type_ (str_):
    if str_.traffic_source == 'yandex' or str_.traffic_source == 'google':
        return 'organic'
    elif (str_.traffic_source == 'paid' or str_.traffic_source == 'email') & (str_.region == 'Russia'):
        return 'ad'
    elif (str_.traffic_source == 'paid' or str_.traffic_source == 'email') & (str_.region != 'Russia'):
        return 'other'
    else:
        return str_.traffic_source
visits['source_type'] = visits.apply(source_type_, axis=1)
visits.head(10)
```
