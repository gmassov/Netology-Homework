# Задание 1
Изучить представленный набор данных на основе описания его столбцов, загрузить его и оставить 8 столбцов для дальнейшего изучения: surgery?, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.

# Задание 3. Работа с пропусками

```python
for col in horse.columns:
    pct_missing = horse[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

1. Можно ничего не менять, т.к. количество пропусков незначительное
2. Пропуски отсутствуют
3. Можно заменить модой, т.к. это наиболее часто встречаемое значение в выборке
4. Пульс заполнить медианой, поскольку пропусков по пульсу не много, но они значимые
5. Аналогично с п.4
6. temp_of_extremities - заполнить модой
7. Аналогично с п.4

```python

horse.rectal_temperature.fillna(horse.rectal_temperature.mode()[0], inplace=True)
```

```python

horse.pulse.fillna(horse.pulse.median(), inplace=True)
```

```python
horse.respiratory_rate.fillna(horse.respiratory_rate.median(), inplace=True)
```

```python
horse.temp_of_extremities.fillna(horse.temp_of_extremities.mode()[0], inplace=True)
```

```python
horse.pain.fillna(horse.pain.median(), inplace=True)
```
